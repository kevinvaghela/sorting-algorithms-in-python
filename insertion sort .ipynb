```python
import random
beg=10
end=10000
lst = []
for i in range(1000):
    values = random.randint(beg, end)
    lst.append(values)

def insertion_sort(lst):
    indexing_range = range(1, len(lst))
    
    for i in indexing_range:
        num_to_sort = lst[i]
        
        while lst[i-1] > num_to_sort and i > 0:
            lst[i], lst[i-1] = lst[i-1], lst[i]
            i = i - 1
            
    return lst
```

```python
%%time
insertion_sort(lst)
```
